```python
import mat73
import numpy as np
import pandas as pd

data_dict = mat73.loadmat('simState.mat')
```

```python
center_freq_sim = mat73.loadmat('simConfigs.mat',only_include='simConfigs/CARRIER_FREQUENCY_IN_MHZ')
center_freq = center_freq_sim['simConfigs']['CARRIER_FREQUENCY_IN_MHZ']
ue_h_sim = mat73.loadmat('simConfigs.mat',only_include='simConfigs/RX_ANT_HEIGHTS_TO_INSPECT_IN_M')
ue_h_all = ue_h_sim['simConfigs']['RX_ANT_HEIGHTS_TO_INSPECT_IN_M']
print(ue_h_all)
```

```python
d = data_dict['simState']
coverage_map = d['coverageMaps']
blockage_map = d['blockageMaps']
ue_xy  = d['mapGridXYPts']
ue_h = ue_h_all[0]
bs_xyh = d['CellAntsXyhEffective']
operator = 0  #assuming we have just one operator for all simulations
```

```python
print((ue_xy[:, [1]]))
print(ue_h)
```

```python
ue_x = np.concatenate(ue_xy[:, [0]])
ue_y = np.concatenate(ue_xy[:, [1]])
blockage = blockage_map[0]
coverage = coverage_map[0]
bs_x = np.concatenate(bs_xyh[:, [0]])
bs_y = np.concatenate(bs_xyh[:, [1]])
bs_h = np.concatenate(bs_xyh[:, [2]])
print(len((bs_x)))
print(len((ue_x)))
#print((blockage))
```

```python
points_ue = list(zip(ue_x, ue_y))
points_bs = list(zip(bs_x, bs_y,bs_h))
print(points_bs)
```

```python
def euclidean_distance(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5


result = []
euclidean_dist = []
for i in points_ue:
    closest_point = None
    smallest_distance = float('inf')
    
    for j in points_bs:
        distance = euclidean_distance(i, (j[0], j[1]))
        
        if distance < smallest_distance:
            smallest_distance = distance
            closest_point = j

    result_point = (closest_point[0], closest_point[1], closest_point[2])
    result.append(result_point)   
    euclidean_dist.append(smallest_distance)     

print(result)
print(len(euclidean_dist))
```

```python
bs_x = [point[0] for point in result]
bs_y = [point[1] for point in result]
bs_h = [point[2] for point in result]
print(bs_x)
```

```python
## pt = 46, gt = 18
#rsrp = pt+gt-pl
rsrp = 46+18-coverage
print(rsrp)
```

```python
df = pd.DataFrame({'operator':operator,
                    'blockage': blockage,
                   'ue_x': ue_x,
                   'ue_y': ue_y,
                    'ue_h':ue_h,
                    'bs_x': bs_x,
                   'bs_y': bs_y,
                    'bs_h':bs_h,
                    'freq':center_freq,
                    'dist':euclidean_dist,
                    'y': coverage,
                    'rsrp':rsrp
                   })
```

```python
print(df)
```
